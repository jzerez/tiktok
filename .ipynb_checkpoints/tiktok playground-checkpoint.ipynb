```python
from scipy import integrate as spi
```

```python
def grav(t, state):
    a = -9.8;
    res = np.zeros_like(state)
    # delta V is equal to gravitational acceleration
    res[0] = a
    # delta X is equal to velocity
    res[1] = state[0]
    return res

def spring(t, state):
    a = -1 * state[1]
    res = np.zeros_like(state)
    # delta V is equal to acceleration
    res[0] = a
    # delta X is equal to velocity
    res[1] = state[0]  + 1*np.sin(t / 2)
    return res



state = np.array([0, 1])

res = spi.solve_ivp(spring, (0, 30), state, max_step=0.01)

print(res)
plt.plot(res.t, res.y[0, :])
plt.plot(res.t, res.y[1, :])
plt.legend(['velocity', 'position'])
```

```python
def spring(t, state):
    k = 1
    m = 1

    omega = (k/m)**0.5
    res = np.zeros_like(state)
    res[0] = -state[1] * k / m
    res[1] = state[0]
    return res

state = np.array([0, 1])


res = spi.solve_ivp(spring, (0, 5), state, max_step=0.01)

plt.plot(res.t, res.y[0, :])
plt.plot(res.t, res.y[1, :])
# plt.legend(['velocity', 'position'])

plt.figure()
plt.plot(res.t)
```

```python
met_mass = 1
num_mets = 25
system_mass = num_mets * met_mass
met_freq = 1 * 2 * np.pi
met_k  = met_freq**2 * met_mass

phases = np.random.random(num_mets) * np.pi * 2
# array that contains the state (velocity and position) of each metronome
states = np.array([1 * met_freq * np.cos(phases),
                        1 * np.sin(phases)]).T

def step(t, states):
        states = np.reshape(states, (len(states)//2, 2))

        res = np.empty_like(states)
        
        met_net_momentum = np.sum(states[:, 0] * met_mass)
        # ensure that momentum of the entire is conserved at zero
        system_vel = met_net_momentum / system_mass
        
        print(system_vel / np.mean(states[:, 0]))
        
        # The change in position is equal to the velocity
        res[:, 1] = -system_vel + states[:, 0]

        # The change in velocity is equal to the acceleration
        res[:, 0] = states[:, 1] / -met_mass * met_k 

        return res.ravel()
res = spi.solve_ivp(step, (0, 10), states.ravel(), max_step=0.005)

print(res.y.shape)
# print(res.t)
print(len(res.t))


# TODO: sum up velocities or positions for all of the metronomes
# see if sync increases over time
vel_inds = np.arange(20)[::2]
vels = res.y[vel_inds, :]

for i in np.arange(20)[::8]:
    plt.plot(res.t, res.y[i, :])
```
